```python
from flask import Flask, render_template, request,jsonify
from flask_cors import CORS,cross_origin
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
```

```python
flipkart_url = "https://www.flipkart.com/search?q=" + "iphone7" #search?q= stands for search query
```

```python
flipkart_url
```

```python
uClient = uReq(flipkart_url) #uReq is an alias of urlopen which requests to open the above flipkart_url
```

```python
uClient
```

```python
flipkartPage = uClient.read()
```

```python
flipkartPage #entire html page I get
```

```python
uClient.close()
```

```python
flipkart_html = bs(flipkartPage, "html.parser") #Beautiful Soup gets rid of the irrelevant info and provides some structure to the html page
```

```python
flipkart_html
```

```python
bigboxes = flipkart_html.findAll("div", {"class": "_1AtVbE col-12-12"}) #each iphone product has a class tage denoted by the 
#one in parentheses, so we want to refer to all the iphone products on that page 
bigboxes
```

```python
del bigboxes[0:3]
```

```python
len(bigboxes) #on the search query page , we have information available of 30 iphone7 products 
```

```python
box = bigboxes[0] #reaching out to the first iphone7 product on the page 
box
```

```python
len(bigboxes)
```

```python
box.div.div.div.a['href']
```

```python
productLink = "https://www.flipkart.com" + box.div.div.div.a['href']
productLink # this gives me the url of the apple-iphone-7-black-128-gb on the iphone7 search query page
```

```python
prodRes = requests.get(productLink) #gives me all the information on the above product link
prodRes.encoding='utf-8'
```

```python
prodRes #request successful
```

```python
prod_html = bs(prodRes.text, "html.parser")
print(prod_html)
```

```python
commentboxes = prod_html.find_all('div', {'class': "_16PBlm"})  # gives me the class tags of all the reviews on the page
```

```python
len(commentboxes) #11 comments are available on that page
```

```python
commentboxes[0]
```

```python
name = commentboxes[0].div.div.find_all('p', {'class': '_2sc7ZR _2V5EHH'})[0].text
name #scrap the name of the firts person of the 1st review
```

```python
rating = commentboxes[0].div.div.div.div.text #scrap the rating of the firts person of the 1st review
rating
```

```python
commentHead = commentboxes[0].div.div.div.p.text
commentHead
```

```python
comtag = commentboxes[0].div.div.find_all('div', {'class': ''})
comtag
```

```python
comtag[0]
```

```python
custComment = comtag[0].div.text
custComment #scrap the comment of the first person of the 1st review
```
